```python
import re
import pandas as pd

# Pastikan kolom 'description' ada di df_specs
if 'description' not in df_specs.columns:
    print("Kolom 'description' tidak ditemukan di DataFrame df_specs.")
else:
    # Identifikasi baris dengan nilai null di 'kamar tidur' atau 'kamar mandi'
    kolom_kamar = ['kamar tidur', 'kamar mandi']
    if all(col in df_specs.columns for col in kolom_kamar):
        null_kamar_indices = df_specs[df_specs['kamar tidur'].isnull() | df_specs['kamar mandi'].isnull()].index

        # Iterasi melalui baris yang memiliki nilai null pada kolom kamar
        for index in null_kamar_indices:
            description_value = df_specs.loc[index, 'description']
            # Check if the returned object is a Series and extract the value if it is
            if isinstance(description_value, pd.Series):
                 # Assuming the Series has only one element if it's from .loc with a single index
                 description = description_value.iloc[0]
            else:
                description = description_value

            if pd.isna(description): # Lewati jika deskripsi juga null
                continue

            description = str(description).lower() # Ubah ke huruf kecil

            # === Pola dan Ekstraksi untuk Kamar Tidur ===
            # Fetch the current value from 'kamar tidur' and ensure it's a scalar
            kamar_tidur_value = df_specs.loc[index, 'kamar tidur']
            if isinstance(kamar_tidur_value, pd.Series):
                kamar_tidur_scalar = kamar_tidur_value.iloc[0]
            else:
                kamar_tidur_scalar = kamar_tidur_value

            # Mencari "kamar tidur:", "kamar tidur ", "bedrooms:", atau "bedrooms " diikuti angka.
            # Handle juga format "+ Angka"
            bedroom_match = re.search(r'(?:kamar\s*tidur|bedrooms)\s*[: ]\s*(\d+)(?:\s*\+\s*(\d+))?', description)
            if bedroom_match:
                try:
                    bedroom_count = int(bedroom_match.group(1))
                    # Cek apakah ada grup penangkap kedua (angka setelah +)
                    if bedroom_match.group(2):
                        bedroom_count += int(bedroom_match.group(2))

                    # Use the scalar value for the isnull check
                    if pd.isna(kamar_tidur_scalar):
                        df_specs.loc[index, 'kamar tidur'] = bedroom_count
                except ValueError:
                    pass # Lewati jika tidak bisa diubah ke integer

            # Fetch the current value from 'kamar tidur' again after potential update and ensure it's a scalar
            kamar_tidur_value_after_match = df_specs.loc[index, 'kamar tidur']
            if isinstance(kamar_tidur_value_after_match, pd.Series):
                 kamar_tidur_scalar_after_match = kamar_tidur_value_after_match.iloc[0]
            else:
                 kamar_tidur_scalar_after_match = kamar_tidur_value_after_match

            # Pola regex alternatif jika tidak ada ":" atau spasi setelah kata kunci, tetapi ada angka sebelum
            # Misalnya "\n4 Kamar Tidur" atau "\n4 Bedrooms"
            # Use the scalar value for the isnull check
            if pd.isna(kamar_tidur_scalar_after_match):
                bedroom_alt_match = re.search(r'(\d+)\s*(?:kamar\s*tidur|bedrooms)', description)
                if bedroom_alt_match:
                     try:
                        bedroom_count_alt = int(bedroom_alt_match.group(1))
                        df_specs.loc[index, 'kamar tidur'] = bedroom_count_alt
                     except ValueError:
                        pass

            # === Pola dan Ekstraksi untuk Kamar Mandi ===
            # Fetch the current value from 'kamar mandi' and ensure it's a scalar
            kamar_mandi_value = df_specs.loc[index, 'kamar mandi']
            if isinstance(kamar_mandi_value, pd.Series):
                 kamar_mandi_scalar = kamar_mandi_value.iloc[0]
            else:
                 kamar_mandi_scalar = kamar_mandi_value

            # Mencari "kamar mandi:", "kamar mandi ", "bathrooms:", atau "bathrooms " diikuti angka.
            # Handle juga format "+ Angka"
            bathroom_match = re.search(r'(?:kamar\s*mandi|bathrooms)\s*[: ]\s*(\d+)(?:\s*\+\s*(\d+))?', description)
            if bathroom_match:
                try:
                    bathroom_count = int(bathroom_match.group(1))
                    # Cek apakah ada grup penangkap kedua (angka setelah +)
                    if bathroom_match.group(2):
                        bathroom_count += int(bathroom_match.group(2))

                    # Use the scalar value for the isnull check
                    if pd.isna(kamar_mandi_scalar):
                         df_specs.loc[index, 'kamar mandi'] = bathroom_count
                except ValueError:
                    pass # Lewati jika tidak bisa diubah ke integer

            # Fetch the current value from 'kamar mandi' again after potential update and ensure it's a scalar
            kamar_mandi_value_after_match = df_specs.loc[index, 'kamar mandi']
            if isinstance(kamar_mandi_value_after_match, pd.Series):
                 kamar_mandi_scalar_after_match = kamar_mandi_value_after_match.iloc[0]
            else:
                 kamar_mandi_scalar_after_match = kamar_mandi_value_after_match

            # Pola regex alternatif jika tidak ada ":" atau spasi setelah kata kunci, tetapi ada angka sebelum
            # Misalnya "\n3 Kamar Mandi" atau "\n3 Bathrooms"
            # Use the scalar value for the isnull check
            if pd.isna(kamar_mandi_scalar_after_match):
                 bathroom_alt_match = re.search(r'(\d+)\s*(?:kamar\s*mandi|bathrooms)', description)
                 if bathroom_alt_match:
                      try:
                        bathroom_count_alt = int(bathroom_alt_match.group(1))
                        df_specs.loc[index, 'kamar mandi'] = bathroom_count_alt
                      except ValueError:
                        pass

        # Tampilkan beberapa baris pertama atau hitung ulang null
        print("\nData df_specs setelah mencoba mengisi nilai null kamar dari deskripsi:")
        display(df_specs.head())

        print("\nJumlah nilai null di df_specs setelah mencoba mengisi dari deskripsi:")
        print(df_specs[kolom_kamar].isnull().sum())
    else:
         print(f"Salah satu kolom kamar ({kolom_kamar}) tidak ditemukan di df_specs.")
```

```python
# Daftar kata kunci untuk setiap fasilitas
keywords = {
    'kolam_renang': ['kolam renang', 'swimming pool', 'kolam'],
    'rumah_sakit_klinik': ['rumah sakit', 'klinik', 'hospital', 'clinic', 'kesehatan', 'rs', 'puskesmas'],
    'tol': ['tol', 'akses tol', 'gerbang tol'],
    'sekolah': ['sekolah', 'school', 'universitas', 'sd', 'smp', 'sma', 'smk', 'univ'],
    'restoran_kuliner': ['restoran', 'kuliner', 'tempat makan', 'restaurant'],
    'pusat_perbelanjaan': ['mall', 'pasar', 'pusat perbelanjaan', 'shopping center', 'market', 'mal'],
    'transportasi': ['bus', 'bis', 'halte', 'angkot', 'kereta', 'krl', 'transportasi', 'terminal','stasiun'] # Tambahkan kata kunci transportasi
}

# Daftar kolom yang akan diperiksa untuk fasilitas (termasuk kolom transportasi)
# Kolom ini akan digunakan untuk mencari semua fasilitas, termasuk transportasi
columns_to_check = {
    'kolam_renang': ['description'],
    'rumah_sakit_klinik': ['description'],
    'tol': ['description'],
    'sekolah': ['description', 'poi_sekolah_text'], # Tambahkan poi_sekolah_text untuk sekolah
    'restoran_kuliner': ['description', 'poi_perbelanjaan_text'], # Tambahkan poi_perbelanjaan_text
    'pusat_perbelanjaan': ['description', 'poi_perbelanjaan_text'], # Tambahkan poi_perbelanjaan_text
    'transportasi': ['description', 'poi_structured_text', 'poi_perbelanjaan_text', 'poi_sekolah_text'] # Kolom untuk transportasi
}

# Buat salinan DataFrame untuk menyimpan hasil dengan kolom fasilitas baru
df_withfasility = df_specs.copy()

# Fungsi untuk memeriksa keberadaan kata kunci dalam teks dengan fuzzy matching di beberapa kolom
def check_keywords_fuzzy_multi_col(row, keyword_list, columns, threshold=85):
    for col in columns:
        if col in row and not pd.isna(row[col]): # Cek apakah kolom ada dan nilainya tidak null
            text = str(row[col]).lower() # Ubah teks menjadi huruf kecil
            for keyword in keyword_list:
                if fuzz.partial_ratio(text, keyword.lower()) > threshold:
                    return 1
                if fuzz.ratio(text, keyword.lower()) > threshold:
                    return 1
    return 0

# Buat kolom baru untuk setiap kategori menggunakan fungsi fuzzy matching di beberapa kolom
for col_name, keyword_list in keywords.items():
    # Ambil daftar kolom yang relevan untuk kategori ini
    relevant_columns = columns_to_check.get(col_name, ['description']) # Default ke description jika tidak ada di columns_to_check

    df_withfasility[col_name] = df_withfasility.apply(
        lambda row: check_keywords_fuzzy_multi_col(row, keyword_list, relevant_columns, threshold=85),
        axis=1 # Terapkan fungsi ke setiap baris
    )

# Tampilkan beberapa baris pertama dari DataFrame baru
print("\nDataFrame dengan kolom fasilitas baru (dengan fuzzy matching, cek di beberapa kolom):")
display(df_withfasility.head(15))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Inisialisasi model Random Forest Regressor
# n_estimators=100 adalah jumlah pohon dalam forest
# random_state=42 untuk hasil yang konsisten
model_rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Melatih model menggunakan data latih
print("Melatih model Random Forest...")
model_rf.fit(X_train, y_train)
print("Model Random Forest selesai dilatih.")

# Melakukan prediksi pada data uji
y_pred_rf = model_rf.predict(X_test)

# # Evaluasi Model Random Forest

# Menghitung metrik evaluasi
mae_rf = mean_absolute_error(y_test, y_pred_rf)
mse_rf = mean_squared_error(y_test, y_pred_rf)
rmse_rf = np.sqrt(mse_rf)
r2_rf = r2_score(y_test, y_pred_rf)

# Menampilkan hasil evaluasi
print("Evaluasi Model Random Forest")
print(f"Mean Absolute Error (MAE): {mae_rf:.2f}")
print(f"Mean Squared Error (MSE): {mse_rf:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse_rf:.2f}")
print(f"R-squared (R2): {r2_rf:.2f}")

# Opsional: Visualisasi prediksi vs aktual
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_rf, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2) # Garis ideal
plt.xlabel("Harga Aktual")
plt.ylabel("Harga Prediksi (Random Forest)")
plt.title("Aktual vs Prediksi (Random Forest)")
plt.grid(True)
plt.show()
```
